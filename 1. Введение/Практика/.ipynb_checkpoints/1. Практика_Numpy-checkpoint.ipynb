10) Создать матрицу случайных чисел размера 10 на 10 и отнять среднее по строкам из каждого элемента строки

```python
matrix = np.random.rand(10, 10)
row_means = np.mean(matrix, axis=1, keepdims=True)
normalized_matrix = matrix - row_means
normalized_matrix
```

12) Создать матрицу случайных чисел 5x5. Добавить к ней столбецы максимумов и минимумов по строкам. Получится матрица 5x7.  

```python
matrix = np.random.rand(5, 5)
row_max = np.max(matrix, axis=1)
row_min = np.min(matrix, axis=1)
matrix_with_max_min = np.column_stack((matrix, row_max, row_min))
matrix_with_max_min
```

13) Вектор чисел от 1 до 51 преобразовать в вектор нечетных чисел и в вектор четных чисел, поменять их размерности, чтобы получились две матрица. Найти произведение этих матриц. Найти обратную матрицу эётого произведения.  

```python
# Создаем вектор чисел от 1 до 51
original_vector = np.arange(1, 52)

# Создаем вектор нечетных чисел
odd_vector = original_vector[original_vector % 2 == 1]

# Создаем вектор четных чисел
even_vector = original_vector[original_vector % 2 == 0]

# Вычисляем максимальную длину векторов
max_length = max(len(odd_vector), len(even_vector))

# Добавляем нули, чтобы сделать оба вектора одной длины
odd_vector = np.pad(odd_vector, (0, max_length - len(odd_vector)), 'constant')
even_vector = np.pad(even_vector, (0, max_length - len(even_vector)), 'constant')

# Преобразуем векторы в квадратные матрицы
odd_matrix = odd_vector.reshape(-1, 1)
even_matrix = even_vector.reshape(-1, 1)

# Вычисляем произведение матриц
result_matrix = np.dot(odd_matrix, even_matrix.T)

# Вычисляем псевдообратную матрицу произведения
pseudo_inverse_result_matrix = np.linalg.pinv(result_matrix)

print(result_matrix)
print(pseudo_inverse_result_matrix)
```

14) Создать 8x8 матрицу и заполнить её в шахматном порядке нулями и единицами.

```python
chessboard_matrix = np.zeros((8, 8), dtype=int)

chessboard_matrix[1::2, ::2] = 1 
chessboard_matrix[::2, 1::2] = 1

chessboard_matrix
```
