## 2. Корреляционная матрица

Напомним, что корреляция отражает взаимосвязь двух случайных величин. Она бывает положительная и отрицательная. Чем ближе коэффициент корреляции к нулю, тем меньше взаимосвязь. Чем больше абсолютная корреляци, тем взаимосвязь больше.

Постройте heatmap корреляционной матрицы. Матрица формируется средствами pandas, со стандартным значением параметров.

**Ответьте на вопрос**: с какими признаками количество прокатов коррелирует меньше всего (около 4 штуки). А с какими больше всего?

Ответ Меньше всего - holiday/workingday/windspeed/weather. Больше всего - registred/casual/temp/atemp

```python
correlation_matrix = df.drop(columns=['datetime']).corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Heatmap')
plt.show()
```

## 3. Barpot

Постройте Bar-график суммарного количества прокатов велосипедов по месяцам за каждый год одновременно. (будет 24 столбика)

**Ответьте на вопрос:** почему в предыдущем графике была такая большая корреляция между количеством прокатов и годом?

Ответ: В предыдущем графике не была большая корреялция между количеством прокатов и годом

```python
df['datetime'] = pd.to_datetime(df['datetime'])

df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month

monthly_counts = df.groupby(['year', 'month'])['count'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.bar(monthly_counts.index, monthly_counts['count'])
plt.xlabel('Месяц и Год')
plt.ylabel('Суммарное количество прокатов')
plt.title('Суммарное количество прокатов велосипедов по месяцам и годам')
plt.xticks(monthly_counts.index, monthly_counts['year'].astype(str) + '-' + monthly_counts['month'].astype(str), rotation=90)
plt.tight_layout()
plt.show()
```

## 7. Боксплот (ящик с усами)

Постройте график распределения (боксплот) количества прокатов велосипедов по месяцам в зависимости от того рабочий это день или нет.

**Ответьте на вопрос:** почему в некоторых месяцах чаще берут велосипеды в будний день, а в некоторые - в выходной.


Ответ:

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='month', y='count', hue='workingday', data=df)
plt.xlabel('Месяц')
plt.ylabel('Количество прокатов велосипедов')
plt.title('Распределение количества прокатов велосипедов по месяцам')
plt.legend(title='Рабочий день', labels=['Выходной', 'Будний день'])
plt.grid(True)
plt.show()
```
